```python
import pandas as pd

import numpy as np

#import sys
#!{sys.executable} -m pip install seaborn
import seaborn as sns

#!{sys.executable} -m pip install matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
clean_data = pd.read_csv('./output/clean_data.csv', index_col=0)
clean_data
```

```python
# Showing the total amount of accidents happened on each country(Top 10).
plt.figure(figsize=(10, 8))

clean_data["Country"].value_counts().head(10).plot.bar(color = "orange");
plt.xticks(rotation=60,ha='right');

plt.title("Top ten countries with highest amount of accidents");
plt.xlabel('Country');
plt.ylabel('Count');
```

```python
# Showing the total amount of accidents happened on each month since 1960.
plt.figure(figsize=(10, 8))

sns.countplot(x = clean_data["Month"], order=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec", "UNKNOWN"]);

plt.xticks(rotation=60,ha='right');

plt.title("Global amount of accidents per month");
plt.xlabel('Months');
plt.ylabel('Count');
```

```python
# Show attacks per month in USA, Australia and South Africa, 

plt.figure(figsize=(18,8))

plt.subplot(1,3,1) # (nrows, ncols, index)
sns.countplot(x = clean_data[clean_data["Country"] == "USA"]["Month"], order=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec", "UNKNOWN"]).set_title('Accidents per month USA');
plt.xticks(rotation=60,ha='right');

plt.subplot(1,3,2) # (nrows, ncols, index)
sns.countplot(x = clean_data[clean_data["Country"] == "AUSTRALIA"]["Month"], order=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec", "UNKNOWN"]).set_title('Accidents per month AUSTRALIA');
plt.xticks(rotation=60,ha='right');

plt.subplot(1,3,3) # (nrows, ncols, index)
sns.countplot(x = clean_data[clean_data["Country"] == "SOUTH AFRICA"]["Month"], order=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec", "UNKNOWN"]).set_title('Accidents per month SOUTH AFRICA');
plt.xticks(rotation=60,ha='right');

#UNKNOWN COLUMN COULD BE OMITTED IN THE VISUALIZATIONS IF WE JUST REMOVE IT FROM "ORDER".
```

```python
# showing if the global accidents since 1960 (divided by months) where fatal or not.
sns.countplot(x=clean_data.Month, hue=clean_data["Fatal (Y/N)"], order=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec", "UNKNOWN"]);
plt.xticks(rotation=60,ha='right');
```

```python
# Ideally wanted to show the Top 10 countries with the highest number of "fatal accidents = Yes", but the code is showing all the Fatal Y/N accidents instead of only the "Y"....

sns.countplot(data = clean_data[clean_data["Fatal (Y/N)"] == "Y"], x = clean_data["Country"], order = clean_data["Country"].value_counts().iloc[:10].index);
plt.xticks(rotation = 60,ha ='right');
```

```python
#showing the Top 10 countries with highest amount of Fatal (Y/N) records.

sns.countplot(x = clean_data.Country, hue = clean_data["Fatal (Y/N)"], order = clean_data["Country"].value_counts().iloc[:10].index);
plt.xticks(rotation = 60,ha = 'right');
```

```python
# showing the total amount of cases since 1960 per year.
sns.histplot(clean_data["Year"]);
```

```python
# Amount of accidents since 1960 on TOP3 Countries.
plt.figure(figsize=(15,5))

plt.subplot(1,3,1) # (nrows, ncols, index)
sns.histplot(clean_data[clean_data["Country"] == "USA"]["Year"]).set_title('USA - Amount of accidents since 1960');

plt.subplot(1,3,2) # (nrows, ncols, index)
sns.histplot(clean_data[clean_data["Country"] == "AUSTRALIA"]["Year"]).set_title('AUSTRALIA - Amount of accidents since 1960');

plt.subplot(1,3,3) # (nrows, ncols, index)
sns.histplot(clean_data[clean_data["Country"] == "SOUTH AFRICA"]["Year"]).set_title('SOUTH AFRICA - Amount of accidents since 1960');
```

```python
# esta grÃ¡fica muestra el numero de casos en funcion del sexo (M or F)
clean_data['Sex'].value_counts().plot(kind='bar');
plt.xlabel('Sex');
plt.ylabel('Count');
```

```python
#en esta grafica mostramos los tipos de accidente.
sns.countplot(data = clean_data, x = "Type")

plt.xticks(rotation=60,ha='right');
plt.title("Type of accidents");
plt.xlabel('Type');
plt.ylabel('Count');
```
